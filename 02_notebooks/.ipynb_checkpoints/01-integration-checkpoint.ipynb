## Normalization

```r
#tmp <- subset(x = healthy_cd45pos, downsample = 1000)
```

```r
healthy_cd45pos <- NormalizeData(healthy_cd45pos, verbose = FALSE)
healthy_cd45neg <- NormalizeData(healthy_cd45neg, verbose = FALSE)

cirrhosis_cd45pos <- NormalizeData(cirrhosis_cd45pos, verbose = FALSE)
cirrhosis_cd45neg <- NormalizeData(cirrhosis_cd45neg, verbose = FALSE)
```

## Dimensionality Reduction

```r
healthy_cd45pos <- healthy_cd45pos %>%
    FindVariableFeatures(., selection.method = "vst", nfeatures = 2000) %>%
    ScaleData(.) %>%
    RunPCA() %>%
    FindNeighbors(dims = 1:30) %>%
    FindClusters() %>%
    RunUMAP(dims = 1:30)

healthy_cd45neg <- healthy_cd45neg %>%
    FindVariableFeatures(., selection.method = "vst", nfeatures = 2000) %>%
    ScaleData(.) %>%
    RunPCA() %>%
    FindNeighbors(dims = 1:30) %>%
    FindClusters() %>%
    RunUMAP(dims = 1:30)

cirrhosis_cd45pos <- cirrhosis_cd45pos %>%
    FindVariableFeatures(., selection.method = "vst", nfeatures = 2000) %>%
    ScaleData(.) %>%
    RunPCA() %>%
    FindNeighbors(dims = 1:30) %>%
    FindClusters() %>%
    RunUMAP(dims = 1:30)

cirrhosis_cd45neg <- cirrhosis_cd45neg %>%
    FindVariableFeatures(., selection.method = "vst", nfeatures = 2000) %>%
    ScaleData(.) %>%
    RunPCA() %>%
    FindNeighbors(dims = 1:30) %>%
    FindClusters() %>%
    RunUMAP(dims = 1:30)

DimPlot(healthy_cd45pos, label = TRUE)
DimPlot(healthy_cd45neg, label = TRUE)
DimPlot(cirrhosis_cd45pos, label = TRUE)
DimPlot(cirrhosis_cd45neg, label = TRUE)
```

## Merging

```r
healthy <- merge(x = healthy_cd45pos, y = healthy_cd45neg)
cirrhosis <- merge(x = cirrhosis_cd45pos, y = cirrhosis_cd45neg)
```

```r
# Prepare for JoinLayers to speed the execution up
join_prep <- function(seurat_obj){
    cell_names = colnames(seurat_obj)
    reordered_metadata = seurat_obj@meta.data[match(cell_names, seurat_obj@meta.data), ]
    rownames(reordered_metadata) = cell_names
    seurat_obj = AddMetaData(seurat_obj, metadata = reordered_metadata)
    seurat_obj = JoinLayers(seurat_obj)
    return(seurat_obj)
}
```

```r
cirrhosis <- join_prep(cirrhosis)
```

```r
healthy <- join_prep(healthy)
```

```r
healthy[["RNA"]] <- JoinLayers(healthy)
cirrhosis[["RNA"]] <- JoinLayers(cirrhosis)
```
